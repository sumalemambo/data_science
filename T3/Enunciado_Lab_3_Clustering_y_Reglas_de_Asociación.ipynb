````python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN
from sklearn import preprocessing
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, silhouette_samples
from yellowbrick.cluster.silhouette import silhouette_visualizer
````

## 1. Bank Customer Churn

Para este problema procederemos de la siguiente forma: utilizaremos modelos de clustering para subdividir a los clientes del banco de acuerdo a sus perfiles, posteriormente veremos el porcentaje de clientes que se quedaron y abandonaron el banco en cada cluster para determinar que clusters tienen a los clientes mas leales y los menos leales. Posteriormente, ajustaremos un clasificador de arbol de decision para determinar a que grupo de "lealtad" pertenece el cliente.

### 1.1 Carga de datos

````python
# Leer CSV con datos de fuga como DataFrame
# *Nota*: ignoraremos aquellas columnas que no aportan info relevante como el id del cliente o nombre
to_drop_columns = ['RowNumber', 'CustomerId', 'Surname']
df = pd.read_csv('datasets/Customer-Churn-Records.csv', usecols=(lambda x: x not in to_drop_columns))
````

````python
df.head()
````

### 1.2 Preprocesamiento de datos

````python
# Transformamos atributos categoricos a One-Hot encoding
# Estos atributos corresponderan al lugar geografico del cliente, el genero y tipo de targeta
onehot_encoding_df = pd.get_dummies(df[['Geography', 'Gender', 'Card Type']])
````

````python
onehot_encoding_df.head()
````

````python
# Concatenamos las columnas categoricas transformadas a One-hot encoding al df
df = pd.concat([df, onehot_encoding_df], axis=1)
````

````python
df.head()
````

````python
# Dropeamos columnas categoricas originales
df = df.drop(labels=['Geography', 'Gender', 'Card Type'], axis=1)
````

````python
df.head()
````

### 1.3 Clustering

#### 1.3.1 KMeans

Una desventaja que tiene *KMeans* en este problema es que la cantidad de clusters es un hiperparametro del modelo. En consecuencia, debemos a priori fijar la cantidad de perfiles en la que queremos subdividir a los clientes del banco a pesar de no saber cuantos perfiles distintos existen.

````python
# Utilizaremos n_clusters = 5
n_clusters = 5
````

````python
# Fiteamos modelo de clustering KMeans con n_clusters = 5
# *Nota: no utilizamos la columna 'Exited' ya que de lo contrario
# estariamos filtrando informacion al modelo de Clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init="auto").fit(
    df[list(set(df.columns) - set('Exited'))])
````

````python
# Para evaluar los clusters utilizamos la metrica de Silhouette coefficient
visualizer = silhouette_visualizer(kmeans, df[list(set(df.columns) - set('Exited'))], isfitted=True)
````

*Silhouette coefficient* es una metrica que nos permite comparar la diferencia de las distancias al cluster mas cercano distinto al asignado e intra-cluster de cada elemento de una muestra. Si *Silhouette coefficient* $< 0$, entonces hay elementos asignados al cluster equivocado. Si *Silhouette coefficient* $ = 0$, entonces hay elementos que estan a la misma distancia de dos clusters. Si *Silhouette coefficient* $ > 0$, entonces los elementos fueron bien asignados. Entre mas cercano a $1$ decimos que la calidad del clustering es mejor.

Es posible notar que todos los clusters producidos por $KMeans$ tienen un *Silhouette coefficient* $> 0.5$ indicando que el algoritmo produjo una segmentacion de los usuarios de calidad aceptable y que no hay segmentos de usuarios que se confundan demasiado.

#### 1.3.2 DBSCAN

Para este problema $DBSCAN$ podria decirse que posee una ventaja sobre $KMeans$ ya que no requiere el numero de clusters a producir como hiperparametro. Por lo cual puede lograr una segmentacion mas natural de los distintos tipos de cliente que tiene el banco

````python
# Parametros de DBSCAN
# Maxima distancia (epsilon)
eps = 1
# Numero minimo de datos x cluster
min_samples = 22
````

````python
# Normalizamos datos antes de entrenar DBSCAN
min_max_scaler = preprocessing.MinMaxScaler()
df_scaled = min_max_scaler.fit_transform(df[list(set(df.columns) - set('Exited'))])
````

````python
# Fiteamos modelo de clustering de DBSCAN
dbscan = DBSCAN(eps=eps, min_samples=min_samples).fit_predict(df_scaled)
````

````python
# Printeamos el numero de clusters encontrados por DBSCAN
print(f'El numero de clusters encontrados por DBSCAN es: {len(np.unique(dbscan))}')
````

````python
# Ya que no podemos printear el silhouette coefficient para cada cluster
# printeamos el silhouette score
print(f'El valor de silhouette score es : {silhouette_score(df_scaled, dbscan)}')
````

Ya que *silhouette score* $> 0$ existe consistencia en los clusters generados por $DBSCAN$, sin embargo, el valor es menor a $0.5$ y por tanto existen algunos clientes que estan mal asignados en sus respectivos perfiles.

### 1.4 Clasificacion

A continuacion entrenamos un arbol de decision para determinar el segmento al cual pertenece un cliente por cada uno de los algoritmos de clustering usados

````python
# Añadimos el perfil asignada a cada cliente por KMeans y DBSCAN al dataframe
df['kmean'] = kmeans.labels_
df['dbscan'] = dbscan
````

#### 1.4.1 Arbol de decision con clusters de KMeans

````python
# Separamos en dataset de entrenamiento y testeo
X_train, X_test, y_train, y_test = train_test_split(
    df[list(set(df.columns) - set(['Exited', 'kmean','dbscan']))], df['kmean'], test_size=0.33, random_state=42)
````

````python
# Ajustamos el clasificador basado en arbol de decision
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
````

````python
# Predecimos el perfil de clientes en dataset de test
y_pred = clf.predict(X_test)
````

````python
# Evaluamos el arbol de decision
print(classification_report(y_test, y_pred))
````

Es posible notar que el arbol clasificador entrenado posee *recalls* $>= 0.99$ indicando que para cada uno de los perfiles de clientes generados $Kmeans$ al menos el $99\%$ de los clientes que tienen ese perfil son clasificados correctamente. Esto nos indica que nuestro modelo de arbol de decision diferencia correctamente los distintos perfiles de cliente. Por otro lado, la metrica de precision es $>= 0.99$ para todos los perfiles lo cual indica que para cada perfil de los clientes clasificados como pertenecientes al perfil al menos el $99\%$ pertenece efectivamente. Esto nos indica que el modelo se confunden poco en las asignaciones.

#### 1.4.1 Arbol de decision con clusters de DBSCAN

````python
# Separamos en dataset de entrenamiento y testeo
X_train, X_test, y_train, y_test = train_test_split(
    df[list(set(df.columns) - set(['Exited', 'kmean','dbscan']))], df['dbscan'], test_size=0.33, random_state=42)
````

````python
# Ajustamos el clasificador basado en arbol de decision
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, y_train)
````

````python
# Predecimos el perfil de clientes en dataset de test
y_pred = clf.predict(X_test)
````

````python
# Evaluamos el arbol de decision
print(classification_report(y_test, y_pred))
````

Es posible notar que el arbol clasificador entrenado para DBSCAN posee *recalls* $>= 0.67$ indicando que para cada uno de los perfiles de clientes generados $DBSCAN$ al menos el $67\%$ de los clientes que tienen ese perfil son clasificados correctamente. De los clientes pertenecientes al perfil 41 tan solo el $67\%$ son clasificados correctamente por el arbol, esto indica un resultado bastante peor que el algoritmo basado en $KMeans$, sin embargo, esto puede deberse a que el soporte para ese perfil es de 6 lo cual es extremadamente bajo. 
Por otro lado, la metrica de precision es $>= 0.70$ para todos los perfiles lo cual indica que para cada perfil de los clientes clasificados como pertenecientes al perfil al menos el $70\%$ pertenece efectivamente. Esto nos indica que el modelo se confunden poco en las asignaciones pero mas que el arbol basado en KMeans.
