### librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from statsmodels.tsa.stattools import acf, pacf, adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
```

```python
df = pd.read_csv("NYPD_Arrests_Data__Historic_.csv")
```

## Filtrar información del dataset

```python
df = df.loc[:, ['ARREST_DATE','LAW_CAT_CD','ARREST_BORO']]
```

```python
df['ARREST_DATE'] = pd.to_datetime(df['ARREST_DATE'])
```

```python
df.info()
```

```python
df['LAW_CAT_CD'].value_counts()
```

```python
df = df.loc[df['LAW_CAT_CD'] != "I"]
```

```python
df['LAW_CAT_CD'].value_counts()
```

## Valores nulos

# NO C que otra tecnica podemos aplica aca depues vemos ke weamonokkk


```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

## One-Hot Encoding a crimenes

```python
dummy_cols = pd.get_dummies(df['LAW_CAT_CD'], prefix='LAW_CAT_CD')
```

```python
dummy_cols
```

```python
df = pd.concat([df, dummy_cols], axis=1)
df.drop('LAW_CAT_CD', axis=1, inplace=True)
```

## separar barrios de la ciudad y agrupar semanalmente

```python
df_S = df.loc[df['ARREST_BORO'] == "S"]
df_S = df_S.drop('ARREST_BORO', axis=1)
```

```python
df_Q = df.loc[df['ARREST_BORO'] == "K"]
df_Q = df_Q.drop('ARREST_BORO', axis=1)
```

```python
df_S = df_S.set_index('ARREST_DATE').resample('W').sum()
```

```python
df_Q = df_Q.set_index('ARREST_DATE').resample('W').sum()
```

```python
df_S_train = df_S[df_S.index <= datetime(2019,12,31)]
df_S_test = df_S[df_S.index > datetime(2019,12,31)]
```

```python
df_Q_train = df_Q[df_Q.index <= datetime(2019,12,31)]
df_Q_test = df_Q[df_Q.index > datetime(2019,12,31)]
```

```python
fig,ax = plt.subplots()
sns.lineplot(data = df_S_train , x = 'ARREST_DATE', y = 'LAW_CAT_CD_F',ax = ax,label = 'F')
sns.lineplot(data = df_S_train , x = 'ARREST_DATE', y = 'LAW_CAT_CD_M',ax = ax,label = 'M')
sns.lineplot(data = df_S_train , x = 'ARREST_DATE', y = 'LAW_CAT_CD_V',ax = ax,label = 'V')
sns.despine()
```

```python
df_S_train['LAW_CAT_CD_F']

for columna in ['LAW_CAT_CD_F','LAW_CAT_CD_M','LAW_CAT_CD_V']:
    analisis = adfuller(df_S_train[columna])
    print(columna)
    print("    p_valor:",analisis[1])
```

```python
a = df_S_train['LAW_CAT_CD_F'] - df_S_train['LAW_CAT_CD_F'].shift(1)
```

```python
analisis = adfuller(a.dropna())
print("    p_valor:",analisis[1])
```

```python
plot_acf(a.dropna(), lags = 52);
plt.xlabel('lags')
plt.ylabel('autocorr')
sns.despine();
```

```python
plot_pacf(a.dropna(), lags = 52);
plt.xlabel('lags')
plt.ylabel('autocorr')
sns.despine();
```

```python
LAGS = 52

fig = plt.figure(figsize = (10, 10))

((ax1, ax2), (ax3, ax4), (ax5, ax6)) = fig.subplots(3, 2)

plot_acf(df_S_train['LAW_CAT_CD_F'], ax = ax1, lags = LAGS, title = "Autocorrelación")
plot_pacf(df_S_train['LAW_CAT_CD_F'], ax = ax2, lags = LAGS, title = "Autocorrelación Parcial")

plot_acf(df_S_train['LAW_CAT_CD_M'], ax = ax3, lags = LAGS, title = "Autocorrelación")
plot_pacf(df_S_train['LAW_CAT_CD_M'], ax = ax4, lags = LAGS, title = "Autocorrelación Parcial")

plot_acf(df_S_train['LAW_CAT_CD_V'], ax = ax5, lags = LAGS, title = "Autocorrelación")
plot_pacf(df_S_train['LAW_CAT_CD_V'], ax = ax6, lags = LAGS, title = "Autocorrelación Parcial")

fig.tight_layout()
```

```python
ar_order = 0
diff_order = 1
ma_order = 4

test_size = 52
#series = df_S_train['LAW_CAT_CD_F']
series = df_S_train['LAW_CAT_CD_F'] - df_S_train['LAW_CAT_CD_F'].shift(1)
series = series.dropna()
rolling_predictions = []
for i in range(test_size):
    train = series[:-(test_size-i)]
    model = ARIMA(train,order = (ar_order, diff_order, ma_order))
    model_fit = model.fit()
    pred = model_fit.forecast(steps=1)
    rolling_predictions.append(pred.values[0])
```

```python
fig,ax = plt.subplots()
sns.lineplot(data = rolling_predictions,ax = ax,label = 'pred')
sns.lineplot(data = df_S_test['LAW_CAT_CD_F'].reset_index(drop=True),ax = ax,label = 'real')
sns.despine()
```

```python
import itertools
p=d=q=range(0,5)
pdq = list(itertools.product(p,d,q))

for param in pdq:
    try:
        model = ARIMA(df_S_train['LAW_CAT_CD_F'],order=param)
        model_fit = model.fit()
        print(param,model_fit.aic)
    except:
        continue
```

```python
fig,ax = plt.subplots()
sns.lineplot(data = df_Q_train, x = 'ARREST_DATE', y = 'LAW_CAT_CD_F',ax = ax,label = 'F')
sns.lineplot(data = df_Q_train, x = 'ARREST_DATE', y = 'LAW_CAT_CD_M',ax = ax,label = 'M')
sns.lineplot(data = df_Q_train, x = 'ARREST_DATE', y = 'LAW_CAT_CD_V',ax = ax,label = 'V')
sns.despine()
```

```python
model = ARIMA(df_S_train['LAW_CAT_CD_F'], order=(2, 0, 4))
model_fit = model.fit()
```

```python
model_fit.summary()
```

```python
predictions = model_fit.predict(start=df_S_test.index[0], end=df_S_test.index[-1])
```

```python
fig,ax = plt.subplots()
sns.lineplot(data = predictions, x = predictions.index, y = predictions,ax = ax,label = 'pred')
sns.lineplot(data = df_S_test, x = 'ARREST_DATE', y = 'LAW_CAT_CD_F',ax = ax,label = 'real')
sns.despine()
```
